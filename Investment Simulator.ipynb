```python
%pylab inline
import numpy as np
import scipy, scipy.stats
import matplotlib.pyplot as plt
import sys
```

```python

# Expense ratio
ER_list = [0.1, 0.35, 0.5, 1.0, 1.5, 2.5]

# Mean Return
Imean_list = [8.5]

# Std. devation of return
Isigma_list = [8.0]

# Length of each simulation
lsim = 20

# Number of simulations
nsim = 10000
```

```python
for Imean in Imean_list:
    for Isigma in Isigma_list:
        for ER in ER_list:
            rtn_series = np.random.normal(1.+(Imean-ER)/100., Isigma/100., (lsim, nsim))
            total_rtn = np.prod(rtn_series, axis=0)
            if True:
                fig = plt.figure()
                ax = fig.add_subplot(111)

                # the histogram of the data
                n, bins, patches = ax.hist(total_rtn, 50, normed=0, facecolor='green', alpha=0.75)
                plt.show()
            print("Imean = %.1f, Isigma = %.1f, ER = %.2f  |  Mean Rtn. = %.2f, Median Rtn. = %.2f, Std. Dev. = %.2f" % (
                    Imean, Isigma, ER,
                    np.mean(total_rtn),
                    scipy.stats.cmedian(total_rtn),
                    scipy.stats.tstd(total_rtn)))
```
